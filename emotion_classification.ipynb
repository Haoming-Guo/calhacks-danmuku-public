```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#from moviepy.editor import VideoFileClip, clips_array, vfx
import seaborn as sns
from datascience import *
import xml.etree.ElementTree as et
from keras.models import load_model
from keras.preprocessing import sequence
import jieba
import json
```

```python
xtree = et.parse("danmuku_horror.xml")
xroot = xtree.getroot()
```

```python
model = load_model("model/review_emotion_classifier.h5")
with open('model/emotion_dict.json', encoding='utf-8') as f:
    index_dict = json.load(f)
```

```python
def getWeight(sentence):
    sen = []
    words = jieba.lcut(sentence)
    if len(words) >= 139:
        words = words[0:139]
    for w in words:
        try:
            sen.append(index_dict[w])
        except:
            sen.append(0)
    input = sequence.pad_sequences(np.array([sen]), maxlen=139)
    output = model.predict(input)
    ret = output[0][1:6]
    return np.array(ret)
        
```

```python
df_cols = ["time", "text"]
rows = []
for node in xroot:
    if node.tag == "d":
        time = float(node.attrib.get("p").split(",")[0])
        txt = node.text
        rows.append({"time": time, "text": txt})

df = pd.DataFrame(rows, columns=df_cols)
df["text"] = df["text"].str.replace(r"([a-zA-Z0-9])\1{3,}", lambda m: m.group(0)[:3])
df["value"] = df["text"].apply(getWeight)

df.head(40)
```

```python
df["second"] = df["time"].apply(int)
df.head()
```

```python
sums = df['value'].groupby(df['second']).apply(lambda arr:np.average(np.array(arr),axis=0))
sums.head()
```

```python
json.dumps([[f.item() for f in arr] for arr in sums])
```
