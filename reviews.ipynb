```python
import requests
import json
```

```python
'''
兽娘动物园2：5345112
Fate/Apo: 6301
刀剑神域：115472




'''
mids = [25402, 28056, 5345112, 6301, 115472, 3451, 10592, 28221390, 132112, 1293, 3450, 1540, 23856, 28002, 24589, 28223067, 23371915, 11890, 12445, 3335618, 10006]
url = "https://bangumi.bilibili.com/review/web_api/short/list?media_id={mid}&folded=0&page_size=100&sort=0"
cursor = "&cursor={cid}"
results = []
```

```python
def getReviews(mid, results):    
    r = requests.get(url.format(mid=mid)).json()
    while r["result"]["total"] > 0:
        for review in r["result"]["list"]:
            if len(review["content"]) <= 25:
                results.append([review["user_rating"]["score"], review["content"]])
        r = requests.get(url.format(mid=mid) + cursor.format(cid=r["result"]["list"][-1]["cursor"])).json()
```

```python
results = []

for mid in mids:
    try:
        getReviews(mid, results)
    except:
        continue
```

```python
results_filtered = []
for i in range(len(results)):
    if results[i][0] == 2:
        d = {}
        d["label"] = 0
        d["content"] = results[i][1]
        results_filtered.append(d)
    elif results[i][0] == 10:
        d = {}
        d["label"] = 1
        d["content"] = results[i][1]
        results_filtered.append(d)
```
